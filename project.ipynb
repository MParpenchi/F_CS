```python
##Question1: Create a single dataframe with the concatenation of all input csv files, adding a column called country
```

```python
import pandas as     ##For working with dataframes
import glob        ##To find files based on a pattern, for example, all *videos.csv files.
import os      ##To work with paths and file names (extract file names, no paths).

csv_files = glob.glob('*videos.csv')     ##Finds all files in the current folder whose name ends with videos.csv
csv_files
```

```python
df_list = []

for file in csv_files:
    filename = os.path.basename(file)   # It just separates the file name from the path
    country_code = filename[:2]         # First two letters = country code (US, CA, FR, ...)
    temp_df = pd.read_csv(file, encoding='latin-1')   ##Converts the CSV file to a DataFrame and stores it in temp_df.
                                                    ##Specify encoding='latin-1' to avoid character errors (especially if there are special/non-UTF-8 characters in the data)
    temp_df['country'] = country_code   ##You add a new column called country to the DataFrame.
    df_list.append(temp_df)    ##add the DataFrame for this file (this country) to df_list. At the end of the loop, df_list is a list of multiple DataFrames
                            ##ignore_index=True causes a new index from 0 to n-1 to be created for the entire DataFrame, ignoring previous indexes.(If you don't specify this, indexes from different files may be duplicated.)
combined_df = pd.concat(df_list, ignore_index=True)   ##It concatenates all the DataFrames in df_list (row-wise concatenation).     ##ignore_index=True causes a new index from 0 to n-1 to be created for the entire DataFrame, ignoring previous indexes.(If you don't specify this, indexes from different files may be duplicated.)
combined_df.head()
```

```python
##Question14: Read all json files with the video categories
```

```python
import pandas as pd
import glob  ##For finding files with pattern
import json  ## for loading json file
import os  ##working with paths
```

```python
json_files = glob.glob('*_category_id.json')
json_files
```

```python
category_dfs = []

for file in json_files:
    country_code = file[:2]  # Extract the 2-letter country code from the file name (e.g. 'US', 'GB', 'FR')
    
    with open(file, 'r', encoding='utf-8') as f:  # Open and load the JSON file . with is a context manager; when the block is finished, the file is automatically closed.
        data = json.load(f)  #json.load(f) converts the JSON contents into a Python structure (usually a dict) and places it in data
    
    df_cat = pd.json_normalize(data['items'])   # Normalize the 'items' list into a flat DataFrame
    
    df_cat['country'] = country_code      # Add a column indicating the country this category file belongs to
    
    category_dfs.append(df_cat)  # Collect the per-country DataFrame

categories_df = pd.concat(category_dfs, ignore_index=True) # Concatenate all category DataFrames (for all countries) into a single DataFrame
categories_df.head()
```

```python
##Question15: For each country, determine how many videos have a category that is not assignable.
```

```python
# Rename columns from the categories JSON to more convenient names
categories_df = categories_df.rename(columns={
    'id': 'category_id',
    'snippet.title': 'category_title',
    'snippet.assignable': 'assignable'
})
# Ensure category_id is numeric so it matches the type in combined_df
categories_df['category_id'] = categories_df['category_id'].astype(int)
# Quick sanity check: inspect key columns in the categories DataFrame
categories_df[['country', 'category_id', 'category_title', 'assignable']].head()
```

```python
# Merge video data with category metadata (assignable flag) per (country, category_id)
videos_with_cat = combined_df.merge(
    categories_df[['country', 'category_id', 'assignable']],
    how='left',
    on=['country', 'category_id']
)
videos_with_cat[['country', 'category_id', 'assignable']].head()
```

```python
## Keep only videos whose category is not assignable
not_assignable_videos = videos_with_cat[videos_with_cat['assignable'] == False]

# For each country, count how many videos have a non-assignable category
not_assignable_counts = (
    not_assignable_videos
    .groupby('country')
    .size()
    .reset_index(name='n_videos_not_assignable')
)

not_assignable_counts
```
